## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data.head(5)
```

```python
bins = [0, 10, 15, 20 , 25, 30, 35, 40, 120]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age_groups"] = pd.cut(purchase_data["Age"],bins, labels=group_names)
purchase_data.head()
```

```python
totalCount = purchase_data.groupby("Age_groups").count()["Purchase ID"]  
```

```python
percentagePlayers = (purchase_data.groupby("Age_groups").count()/780*100)["SN"]
```

```python
ageDemographicsDataFrame= pd.DataFrame({
    "Total Count": totalCount,
    "Percentage of Players" : percentagePlayers
})
```

```python
ageDemographicsDataFrame
```
