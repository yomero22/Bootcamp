```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_to_load = pd.read_csv("city_data.csv")
ride_data_to_load = pd.read_csv("ride_data.csv")


# Combine the data into a single dataset

pyber_df = pd.merge (city_data_to_load,ride_data_to_load,on="city")

# Display the data table for preview

pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python

ridesbycityu = pyber_df[pyber_df["type"] == "Urban"].groupby("city").count()["ride_id"]
ridesbycitys = pyber_df[pyber_df["type"] == "Suburban"].groupby("city").count()["ride_id"]
ridesbycityr = pyber_df[pyber_df["type"] == "Rural"].groupby("city").count()["ride_id"]

driversbycityu = pyber_df[pyber_df["type"] == "Urban"].groupby("city").count()["ride_id"]
driversbycitys = pyber_df[pyber_df["type"] == "Suburban"].groupby("city").count()["ride_id"]
driverscityr = pyber_df[pyber_df["type"] == "Rural"].groupby("city").count()["ride_id"]

average_fareu = pyber_df[pyber_df["type"] == "Urban"].groupby("city").mean()["ride_id"]
average_fares = pyber_df[pyber_df["type"] == "Suburban"].groupby("city").mean()["ride_id"]
average_farer = pyber_df[pyber_df["type"] == "Rural"].groupby("city").mean()["ride_id"]


# Obtain the x and y coordinates for each of the three city types

#Urban
UxAx = ridesbycityu
UyAx = driversbycityu
UzAx = average_fareu

#Suburban
SxAx = ridesbycitys
SyAx = driversbycitys
SzAx = average_fares

#Rural
RxAx = ridesbycityr
RyAx = driverscityr
RzAx = average_farer


# Build the scatter plots for each city types

UBubble = plt.scatter(UyAx, UxAx, s=UzAx*10, marker="o", facecolors="blue", edgecolors="black", alpha=0.5, label="Urban")
SBubble = plt.scatter(SyAx, SxAx, s=SzAx*10, marker="o", facecolors="purple", edgecolors="black", alpha=0.5, label="Suburban")
RBubble = plt.scatter(RyAx, RxAx, s=RzAx*10, marker="o", facecolors="orange", edgecolors="black", alpha=0.5, label="Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

# Create a legend
Lege = plt.legend(handles=[UBubble, SBubble, RBubble], title="City Types", fontsize=8, loc="best")
Lege.legendHandles[0]._sizes = [50]
Lege.legendHandles[1]._sizes = [50]
Lege.legendHandles[2]._sizes = [50]

# Incorporate a text label regarding circle size
plt.text(45,35,"Note: \nCircle size is equal to driver count per city.")

# Save Figure
plt.savefig("./RDpyber_starer4.png")

```

```python
# Show plot
plt.show()
```
